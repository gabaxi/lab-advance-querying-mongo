```python
# import pymongo to connect Python with MongoDB
from pymongo import MongoClient
# import others
import pprint
import pandas as pd
import numpy as np

# connecting on default host and port
client = MongoClient ('localhost', 27017)

# getting the database
db = client['companies']

# getting the collection
companies = db['companies']
```

### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.

```python
query2 = companies.find({'number_of_employees':{'$gt':5000}}).sort('number_of_employees', -1).limit(20)
for company in query2:
    print(company['name'],':', company['number_of_employees'],'employees')
```

```python
query2 = companies.find({'number_of_employees':{'$gt':5000}}).sort('number_of_employees', -1).limit(20)
query2 = pd.DataFrame(data=query2, columns=['name','number_of_employees'],index=None)
print('top',len(query2),'companies that have more than 5000 employees')
display(query2)
```

### 4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the `name` and `ipo` fields.

```python
query4 = companies.find({'$and': [{'ipo.valuation_amount':{'$gt': 100000000 }},{'founded_year':{'$lt': 2010 }}]},{ 'name': 1,'ipo.valuation_amount': 1, '_id': 0 })
for company in query4:
    print(company['name'],'valuated', company['ipo']['valuation_amount'])
```

### 5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
query5 = companies.find({'$and': [{'number_of_employees':{'$lt': 1000 }},{'founded_year':{'$lt': 2005 }}]}).sort('number_of_employees', -1).limit(10)
query5 = pd.DataFrame(data=query5, columns=['name', 'number_of_employees'],index=None)

print('Top',len(query5),'companies that have less than 1000 employees and have been founded before 2005.')
display(query5)
```

### 6. All the companies that don't include the `partners` field.

### 7. All the companies that have a null type of value on the `category_code` field.

### 8. All the companies that have at least 100 employees but less than 1000. Retrieve only the `name` and `number of employees` fields.
